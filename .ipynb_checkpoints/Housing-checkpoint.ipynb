```python
import os, tarfile, urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    # Create a housing/datasets directory
    os.makedirs(housing_path, exist_ok=True)
    
    # Download the housing.tgz file
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    
    # Extract the housing.csv file from the tgz archive, placing it in the housing directory
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

```python
fetch_housing_data()
```

```python
import pandas as pd

# This function returns a pandas DataFrame object containing the housing data 
def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path)
```

```python
# We now want to choose a test set, containing 20% of all instances in our dataset. 
# The following function is a rather naive approach, which relies on numpy's ability to generate a 
# random permutation. Setting the random seed in advance makes this approach stable, but only as long as our
# dataset does not change!

import numpy as np

def split_train_test(data, test_ratio):
    np.random.seed(42)
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(housing, 0.2)
```

```python
# To choose (test_ratio=)20% of all datapoints for our test set, compute a hash of each instance's identifier.
# The crc32 algorithm returns a 32 bit integer - an instance should belong to our hash set if its identifier
# lies in the according portion of the (0, 2**32) range. This way, results don't vary between runs. 

from zlib import crc32

def test_set_check(identifier, test_ratio):
    return crc32(np.int64(identifier)) < test_ratio * 2**32
```

```python
# This function applies the test_set_check function to all instances in a dataset, and then returns 
# 2 lists: The train set (of elements that "failed" the check, i.e. returned false) and the test set.
def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
# We could use the row index of each instance as its id.
# This is simple, but only works if no row is ever deleted and new data is always appended to the end of the data.

housing_with_id = housing.reset_index()   # adds an 'index' column to our dataset
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, 'index')
```

```python
# Instead, we can count on a district's longitude and latitude being stable.
# However, since the location information is quite coarse, we are introducing some sampling bias, since
# many neighbouring districts will receive the same hash and thus be guaranteed to end up in the same set. 

housing_with_id["id"] = housing["longitude"] * 1000 + housing["latitude"]
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "id")
```

```python
# Simpler still is the following approach to choosing a test set:
# We can use Scikit-Learn's train_test_split function, which does pretty much the same thing as our 
# split_train_test function, with a couple of extra features: The random seed can be set as 
# a parameter, and you can pass it multiple datasets with the same number of rows that are then split on the 
# same indices. That is very useful, e.g. if we have a separate DataFrame for labels.

from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
```

```python
# By choosing one of the previous (purely random) sampling methods, we risk introducing a significant
# sampling bias, especially if our dataset is not very big, relative to our number of attributes. 
# If we want our test set to be representative of all categories of a relevant attribute, this is called
# "stratified sampling".

# In our case, we can categorize the median income of each district, taking note of this information in 
# the "income_cat" column. 
housing["income_cat"] = pd.cut(housing["median_income"], 
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])
```

```python
housing["income_cat"].hist()
```

```python
# Now we can do stratified sampling based on the income category
from sklearn.model_selection import StratifiedShuffleSplit

sss = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 42)
# Calling the split function of the sss object generates two numpy arrays of indices
for train_index, test_index in sss.split(housing, housing["income_cat"]):
    # We only iterate through this for-loop once, using pd's DataFrame.loc function to retrieve two
    # lists of districts, based on the two arrays of indices we generated
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
# We compare the income category proportions in the test set and total dataset, to confirm we were successful. 

print(strat_test_set["income_cat"].value_counts() / len(strat_test_set))

print(housing["income_cat"].value_counts() / len(housing))
```

```python
# We can now remove the income_cat attribute to return the data to its original state

for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
```

```python
# We create a copy of the training set, so that we cannot cheat by looking at the test data & so that we can 
# play around with this copy without causing any harm

housing = strat_train_set.copy()
```

```python
# A 4th option is to use Scikit-Learn's SimpleImputer class to take care of missing values

from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")

# The median can only be computed on numerical attributes, so we drop the text attribute ocean_proximity
housing_num = housing.drop("ocean_proximity", axis=1)

# Now we can fit the imputer to the training data
imputer.fit(housing_num)

# The median of each attribute is stored in the imputer's statistics_ attribute

print(imputer.statistics_)
print(housing_num.median().values)
```

```python
# We can now let the imputer replace missing values with the learned medians

X = imputer.transform(housing_num)

# We can then put this NumPy array back into a pandas DataFrame

housing_tr = pd.DataFrame(X, columns=housing_num.columns, index=housing_num.index)
```

```python
# Let's now handle text/categorical attributes... in this case, just the "ocean_proximity" column.

# DataFrame[ [column names] ] is syntax for extracting the named columns to a new DataFrame.
# Note: DataFrame[column name] returns the specified column as a pandas series. 
housing_cat = housing[["ocean_proximity"]]
housing_cat.head(10)
```

```python
# We can convert these categories from text to numbers to make it easier for our ML algorithm to work with them.
# To this end, we can use Scikit-Learn's OrdinalEncoder class. 

from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
housing_cat_encoded = ordinal_encoder.fit_transform(housing_cat)
housing_cat_encoded[:10]
```

```python
# However, this causes a problem: An ML algorithm may assume that e.g. the categories 0 and 1 are much more
# similar than the categories 0 and 4... which is clearly false. 

# To avoid this problem, we can instead create one binary attribute per category.
# This is called "one-hot encoding". The new attributes are sometimes called "dummy attributes".

from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()
housing_cat_1hot = cat_encoder.fit_transform(housing_cat)
housing_cat_1hot
```

```python
# Transformers in Scikit-Learn feature fit, transform and fit_transform methods. 
# We need to create custom ones for tasks such as certain cleanup operations or combining attributes. 

from sklearn.base import BaseEstimator, TransformerMixin

# We declare the column indices of the attributes needed for our computations
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

# We add 2 base classes to our transformer:
# BaseEstimator - giving us get_params() and set_params(); useful for hyperparameter tuning
# TransformerMixin - giving us the fit_transform() method for free
class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    
    # We include an add_bedrooms_per_room parameter to quickly decide whether this attribute is useful
    def __init__(self, add_bedrooms_per_room = True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
        
    def fit(self, X, y=None):
        return self # Nothing else to do
    
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix] # Element-wise division
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
        
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
# The np.ndarray housing_extra_attribs contains attributes for rooms/household and population/household
housing_extra_attribs = attr_adder.transform(housing.values)
```

```python
# We use Scikit-Learn's Pipeline class to execute the 3 transformations
# - Imputation for missing values
# - Adding combined attributes
# - Feature Scaling (Standardization)
# in the right order. 

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# All but the last estimators in a Pipeline instance must be transformers (with a fit_transform() method)
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("attribs_adder", CombinedAttributesAdder()),
    # Using Standardization, we subtract mean & divide by std (less affected by outliers than normalization)
    ("std_scaler", StandardScaler()) 
])

# The Pipeline calls fit_transform() sequentially on all estimators, passing the output of each call 
# as the parameter of the next cell, unless the last cell is reached, where the pipeline's method we use
# is performed. In this case, we use fit_transform() because the last cell also contains a transformer. 
housing_num_tr = num_pipeline.fit_transform(housing_num)
```

```python
# It would of course be neat to have a single transformer that can handle numerical & categorical attributes. 
# To this end, we can use the ColumnTransformer class

from sklearn.compose import ColumnTransformer

# Get lists of all numerical/categorical attributes
num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

# We string together our "numerical pipeline" and a One Hot Encoder, to create a pipeline which deals with
# all columns of our data. 
# Note: num_pipeline returns a dense, OneHotEncoder a sparse matrix. The sparse_threshold parameter 
# determines at what "density" full_pipeline should return a dense matrix. 0.3 is the default
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
    ], 
    sparse_threshold = 0.3)

housing_prepared = full_pipeline.fit_transform(housing)
```

```python
# After getting, exploring, sampling and preparing the data, we can now select and train a model.
# Let's start with a linear regression model. 

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(housing_prepared, housing_labels)
```

```python
# Let's try our model on some instances from the training set

some_data = housing.iloc[:5]
some_labels = housing_labels.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)
print("Predictions:", lin_reg.predict(some_data_prepared))
print("Labels:", list(some_labels))
```

```python
# Let's measure the regresson model's RMSE (Root Mean Squared Error) on the whole training set

from sklearn.metrics import mean_squared_error
housing_predictions = lin_reg.predict(housing_prepared)
lin_mse = mean_squared_error(housing_labels, housing_predictions)
lin_rmse = np.sqrt(lin_mse)

lin_rmse
```

```python
# A typical prediction error of almost 69k isn't great. We are underfitting the data. 
# Let's try a more complex model - a DecisionTreeRegressor

from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(housing_prepared, housing_labels)
```

```python
# We measure the RMSE on the training set.

housing_predictions = tree_reg.predict(housing_prepared)
tree_mse = mean_squared_error(housing_labels, housing_predictions)
tree_rmse = np.sqrt(tree_mse)

tree_rmse
```

We are probably overfitting by quite a bit!

```python
# To evaluate the decision tree model in a more meaningful way, we use Scikit-Learn's K-fold cross validation
# feature, splitting the training data into 10 folds, training and evaluating the model 10 times, using each
# fold 9 times for training and once for testing. 

from sklearn.model_selection import cross_val_score 
scores = cross_val_score(tree_reg, housing_prepared, housing_labels, scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
# We write a function to inspect the results 

def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
    
display_scores(tree_rmse_scores)
```

These results actually look kinda bad, far from the "perfect" result we got when evaluating on the training set as a whole.

```python
# Let's evaluate our linear regression model in the same way

lin_scores = cross_val_score(lin_reg, housing_prepared, housing_labels,
                             scoring="neg_mean_squared_error", cv=10)

lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

The (underfitting) Linear Regression model actually performs slightly better than the (overfitting) decision tree model. But they both are not doing very well.
